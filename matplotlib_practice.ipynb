```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]


# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.2, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# Create Line Chart using the MATLAB Method

```python
# Create the plot
# chain the plot() function to plt. Inside the parentheses of the plot() function, add the x_axis and y_axis parameters

plt.plot(x_axis, y_axis)
```

# Create Line Chart using Object Oriented Method

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

```

## Another way to create the same chart using the Object Oriented Method

```python
# The fig, ax = plt.subplots() line of code can also be written like this.
#Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)


```

### Another quick way to create the line chart with the Object Oriented Method

```python
# Using the object oriented method to: Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

# Adding color, line width, and a marker unsing Matplolib

```python
# Create the plot, add a label for the legend, adding a line color, line width, and a marker for our data points.
plt.plot(x_axis, y_axis, label='Boston')
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

### Skill Drill

Using the object-oriented approach, make the following changes to your line chart:

• Change the line color to green.

• Change the marker to a diamond.

• Change the line width to 2 points.

• Add a legend for the city of Boston.

• Add a title and axes labels.

• Add a y-axis limit.

• Add grid lines.

```python
# Create the plot using the object-oriented approach.

ax = plt.axes()
#1.Change the line color to green.
#2.Change the marker to a diamond.
#3.Change the line width to 2 points.
#4.Add a legend for the city of Boston.
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label='Boston')

ax.legend()

#5.Add a title and axes labels.
ax.set_title("PyBer Fare by Month")

ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

#6.Add a y-axis limit.
ax.set_ylim(0, 45)

#7.Add grid lines.
ax.grid()

```
